```python
# 1. Read Total profit of all months and show it using a line plot
total_profit_month=df.groupby("month_number")["total_profit"].sum()
total_profit_month
plt.plot(total_profit_month)
```

```python
# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3
total_profit_month_units=df.groupby("month_number")["total_units"].sum()
plt.plot(total_profit_month_units,color="red",linestyle="dotted",marker="o",linewidth=3)
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
```

```python
# 3. Read all product sales data and show it  using a multiline plot
#     square marker with color=green 

plt.plot(df["month_number"],df["facecream"])
plt.plot(df["month_number"],df["facewash"])
plt.plot(df["month_number"],df["toothpaste"])
plt.plot(df["month_number"],df["bathingsoap"])
plt.plot(df["month_number"],df["shampoo"])
plt.plot(df["month_number"],df["moisturizer"])
plt.xlabel("Month Number")
```

```python
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
for p in products:
    plt.plot(df["month_number"],df[p],label=p,marker="o",markerfacecolor="green")
plt.legend()
```

```python
# 4.Read toothpaste sales data of each month and show it using a scatterplot
plt.scatter(df["month_number"],df["toothpaste"])
```

```python
# 5.Read face cream and facewash product sales data and show it using the bar chart with grid line
plt.bar(df["month_number"],df["facecream"],label="facecream")
plt.bar(df["month_number"],df["facewash"],label="facewash")
plt.grid()
```

```python
# 6.Read the total profit of each month and show it using the histogram to see the most common profit ranges
total_profit_of_each_month=df.groupby("month_number")["total_profit"].sum()
total_profit_of_each_month
plt.hist(total_profit_of_each_month,edgecolor="black")
```

```python
# 7.Calculate total sale data for last year for each product and show it using a Pie chart
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
sale_summary={p:df[p].sum() for p in products}
plt.pie(sale_summary.values(),labels=products,autopct="%1.1f%%")
```

```python
# 8. Read Bathing soap facewash of all months and display it using the Subplot
figure,axes=plt.subplots(1,2)
axes[0].plot(df["month_number"],df["facewash"])
axes[1].plot(df["month_number"],df["bathingsoap"])
```
