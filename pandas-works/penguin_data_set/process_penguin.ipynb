```python
# fliper length body mass line chart
fliper_lenght_body_mas=df.groupby("flipper_length_mm")["body_mass_g"].mean()
fliper_lenght_body_mas
```

```python
plt.plot(fliper_lenght_body_mas)
```

```python
bill_lenght_body_mas=df.groupby("bill_length_mm")["body_mass_g"].mean()
bill_lenght_body_mas
plt.plot(bill_lenght_body_mas)
```

```python
bill_depth_body_mass=df.groupby("bill_depth_mm")["body_mass_g"].mean()
bill_depth_body_mass
plt.plot(bill_depth_body_mass)
```
