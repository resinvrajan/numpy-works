```python
courses=["django","datascience","testing","mearn"]
students_count=[24,20,15,18]
students_count1=[23,54,20,12]
plt.plot(courses,students_count,marker="^",linestyle="solid",color="#54a832",linewidth=2)
plt.plot(courses,students_count1,color="red",marker=".")
plt.title("luminar tsr",color="blue",size=24,weight="bold")
plt.xlabel("courses",color="skyblue",weight="bold")
plt.ylabel("student_count",color="green",weight="bold")
plt.show()
```

```python
courses=["django","datascience","testing","mearn"]
students_count=[24,20,15,18]
plt.pie(students_count,labels=courses,autopct="%1.1f%%",shadow=True)
```
